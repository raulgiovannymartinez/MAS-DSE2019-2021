(12) write a program to find solutions of the 8 queen problem.
>The eight queens problem is the problem of placing eight queens on an 8×8 chessboard such that none of them attack one >another (no two are in the same row, column, or diagonal). Below is one of the arrangements of an 8 queen solution

```python
#import libraries
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

######################################## Functions ########################################

#function to define an empty board
def Get_board_grid(N):
    board = np.zeros([N,N])
    board[:] = np.nan
    return board

#check attack diagonally in all 4 directions
def Check_Diagonal_attack(row,col,board,direction):
    count = 0
    while True:
        try:   
            #check if true or 1
            if board[row,col] == 1:
                count+=1
            #check other diagonal locations based on direction
            if direction == 'top-right':
                row-=1
                col+=1
            elif direction == 'top-left':
                row-=1
                col-=1
            elif direction == 'down-right':
                row+=1
                col+=1
            elif direction == 'down-left':
                row+=1
                col-=1
            #break prevent index from getting negative
            if (row<0) or (col<0):
                break
        except IndexError:
            break
    #return number of attacks found
    return count

#check if the queen is under attack
def Check_under_attack(row,col,board):
    #check rows and cols
    if sum(board[row,:] == True) > 1:
        return True
    elif sum(board[:,col] == True) > 1:
        return True
    #check diagonally 
    diagonal_counts = [Check_Diagonal_attack(row,col,board,'top-right'),Check_Diagonal_attack(row,col,board,'top-left'),
                       Check_Diagonal_attack(row,col,board,'down-right'),Check_Diagonal_attack(row,col,board,'down-left')]
    if sum(diagonal_counts) > 4:
        return True
    #return not in attack or false if the above is not true
    else:
        return False
        
######################################## Main ########################################
    
#define board 
N = 8
Board = Get_board_grid(N)

#define dictionary to store queens position
dict_Q = {}
for i in range(1,N+1):
    dict_Q[i] = [i-1,np.nan]

#initialize variables
solutions = []
Q = 1
col = 0

#loop to find all possible solutions
while True:  
    try:
        #loop to find solution with backtracking strategy
        while Q<len(dict_Q)+1:
            try:
                #find column where queen is not under attack
                while True:
                    #check each column by adding the queen to the board and updating the dictionary
                    dict_Q[Q][1] = col #update dictionary
                    Board[dict_Q[Q][0],dict_Q[Q][1]] = True #update board
                    #if the queen is not under attack continue to the next queen
                    if Check_under_attack(dict_Q[Q][0],dict_Q[Q][1],Board) == False:
                        col = 0 #start with column 0 again
                        Q+=1 #move to next queen
                        break #Good, continue to next queen
                    else:
                        #if under attack, check next column, move out queen from board and reset dictionary information
                        Board[dict_Q[Q][0],dict_Q[Q][1]] = np.nan #update board
                        dict_Q[Q][1] = np.nan #update dictionary
                        col+=1 #move to next column
            except IndexError: #loop until we reach the end of the column
                dict_Q[Q][1] = np.nan #reset current queen position
                Q-=1 #BACKTRACKING
                col = dict_Q[Q][1]+1 #move to next column
                dict_Q[Q][1] = np.nan #reset previous queen position to find new solution
                Board[Q-1,:] = np.nan #reset board from previous queen position
                
        #make a copy of the board for saving them on a list
        copy_Board = np.copy(Board)
        
        #only save boards that are unique 
        if len(solutions) == 0:
            solutions.append(copy_Board)
        elif np.sum(np.equal(Board,solutions[0])) == N:
            print('yes')
        else:
            solutions.append(copy_Board)
            
        #update previous board to start a new solution find
        Q-=1 #BACKTRACKING
        col = dict_Q[Q][1]+1 #move to next column
        dict_Q[Q][1] = np.nan #update dictionary
        Board[Q-1,:] = np.nan #update board
    #continue until there is a key error to exit the loop
    except KeyError:
        break
    
#print one solution as an example in a heatmap
example_board = np.copy(solutions[0])    
example_board = np.nan_to_num(example_board)
ax = sns.heatmap(example_board, linewidth=0.5, cbar = False)
plt.show()
```

(13) Write a program to make a list of all prime numbers less than 3000. A prime number can only be divided evenly by 1 and itself.

```python
#define list to save prime numbers
prime_num = []
N_max = 3000

#scan from 0 to 3000
for number in range(0,N_max + 1):  
   if number > 1: #1 is not prime, start after 1
       for i in range(2,number): #verify is not divisible by other numbers
           if (number % i) == 0:  
               break #not prime
       else:  #append prime numbers
           prime_num.append(number)

#print prime numbers
print(prime_num)
```
