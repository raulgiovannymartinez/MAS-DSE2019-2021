### Q1: Create the dataframes

Download and parse the data files and create the following three pandas dataframes: (your dataframes should have data in the format shown below)

 * US_stat: statistics for the whole United States.
 * states_stat: Statistics for each state.
 * county_stat: Statistics for each county.

```python
import pandas as pd
import urllib
import requests
import numpy as np

#function to check if url path exists
def exists(path):
    r = requests.head(path)
    return r.status_code == requests.codes.ok

#function to parse data using documentation layout
def Get_Data_withLayoutDict(file,layout_dict):
    data = []
    for line in file:
        try:
            #convert byte line to string
            line = line.decode('ascii')
        except AttributeError:
            pass
        
        cols = []
        for key in layout_dict.keys():
            cols.append(line[layout_dict[key][0]-1:layout_dict[key][1]])
        
        data.append(tuple(cols))
    return data

#dictionary to define documentation layout
layout_dict = {'State FIPS':[1,2], 'County FIPS':[4,6],'Poverty Estimate All Ages':[8,15],
        'Poverty Percent All Ages': [35,38],'Poverty Estimate Under Age 18':[50,57],
        'Poverty Percent Under Age 18':[77,80],'Poverty Estimate Ages 5-17':[92,99],
        'Poverty Percent Ages 5-17':[119,122],'Median Household Income':[134,139],
        'Name':[194,238],'Postal':[240,241]}

#generic link for all files
link = 'https://www2.census.gov/programs-surveys/saipe/datasets/%s/%s-state-and-county/'

#define empty dataframe
agg_stat = pd.DataFrame(columns = layout_dict.keys())

#loop over every year
for yr in range(1993,2016+1):
    for file_ in ['est%sall.txt','est%sall.dat','est%sus.txt','est%sus.dat']:
    
        #define url
        url = (link+file_)%(yr,yr,str(yr)[-2:]) 

        #check if file exists for each condition .dat and .txt
        if exists(url):
            
            #to read .txt file data use urllib library
            if '.txt' in file_:
                file = urllib.request.urlopen(url)
            #to read .dat file data use requests library
            elif '.dat' in file_:
                file = requests.get(url).text.split('\n')
            #continue to see if other files are present
            else:
                continue
                
            #get data in a list of tuples, use dictionary to get each corresponding position along the line string
            data = Get_Data_withLayoutDict(file,layout_dict)
            
            #create dataframe,add year, change FIPS format, and add FIPS column
            df_temp = pd.DataFrame(data, columns = layout_dict.keys())
            df_temp['County FIPS'] = [str(i.strip()).zfill(3) for i in df_temp['County FIPS']]
            df_temp['year'] = str(yr)
            df_temp['FIPS'] = [x+y for x,y in zip(df_temp['State FIPS'],df_temp['County FIPS'])]
            
            #aggregate dataframes and exit loop
            agg_stat = pd.concat([agg_stat,df_temp],sort=True)
            break
        
        #try to see if next file on list is present
        else:
            
            continue

#reset index and drop rows where Postal is ''
agg_stat = agg_stat.reset_index(drop=True)
agg_stat.drop(agg_stat[agg_stat.Postal==''].index, inplace=True)
```

### Q4: plot poverty statistics by county

Using the vincent library and the dataframe `county_stat`, generate the following three maps.

1. Overall percentage of poverty for each county in 2000.
![poverty2000](Poverty2000.jpg)
1. Overall percentage of poverty for each county in 2016.

1. Ratio between percentage of change in poverty from 2000 to 2016 for each county. Divided by the nation-wide change in poverty (find this value).

```python
#Get dataframe together with us_countries.topo.json FIPS values

import json
import vincent

with open('us_counties.topo.json','r') as f:
    get_id = json.load(f)
    
#A little FIPS code type casting to ensure keys match
new_geoms = []
for geom in get_id['objects']['us_counties.geo']['geometries']:
    geom['properties']['FIPS'] = int(geom['properties']['FIPS'])
    new_geoms.append(geom)
    
get_id['objects']['us_counties.geo']['geometries'] = new_geoms

with open('us_counties.topo.json', 'w') as f:
    json.dump(get_id, f)
    
#Grab the FIPS codes and load them into a dataframe
geometries = get_id['objects']['us_counties.geo']['geometries']
county_codes = [x['properties']['FIPS'] for x in geometries]
county_df = pd.DataFrame({'FIPS': county_codes}, dtype=str)
county_df = county_df.astype(int)

#cast county_stat dataframe FIPS to int for consistency
county_stat_vincent = county_stat
idx = county_stat_vincent.index
county_stat_vincent.index = county_stat.index.set_levels([idx.levels[0],idx.levels[1].astype(int),idx.levels[2]])
county_stat_vincent = county_stat_vincent.reset_index(level=[1,2])

#Perform an outer join, pad NA's with data from nearest county
merged = pd.merge(county_stat_vincent, county_df, on='FIPS', how='outer')
merged = merged.fillna(method='pad')

#convert 'Poverty Percent All Ages' column to float
merged['Poverty Percent All Ages'] = [float(i) for i in merged['Poverty Percent All Ages']]
```

```python
#1. Overall percentage of poverty for each county in 2000.

from vincent import ValueRef
vincent.core.initialize_notebook()

#define the year
year = '2000'

#set-up the map
county_topo = r'us_counties.topo.json'
geo_data = [{'name': 'counties',
             'url': county_topo,
             'feature': 'us_counties.geo'}]

#plot the map
vis = vincent.Map(data=merged[merged.year==year], geo_data=geo_data, scale=1100,
                  projection='albersUsa', data_bind='Poverty Percent All Ages',
                  data_key='FIPS', map_key={'counties': 'properties.FIPS'})
vis.marks[0].properties.enter.stroke_opacity = ValueRef(value=0.5)

#add format to map
vis.rebind(column='Poverty Percent All Ages', brew='YlGnBu')
vis.scales['color'].type = 'threshold'
vis.scales['color'].domain = [0, 4, 6, 8, 10, 12, 20, 30]
vis.legend(title='Poverty 2000 (%)')
```

```python
#2. Overall percentage of poverty for each county in 2016.

from vincent import ValueRef
vincent.core.initialize_notebook()

#define the year
year = '2016'

#set-up the map
county_topo = r'us_counties.topo.json'
geo_data = [{'name': 'counties',
             'url': county_topo,
             'feature': 'us_counties.geo'}]

#plot the map
vis = vincent.Map(data=merged[merged.year==year], geo_data=geo_data, scale=1100,
                  projection='albersUsa', data_bind='Poverty Percent All Ages',
                  data_key='FIPS', map_key={'counties': 'properties.FIPS'})
vis.marks[0].properties.enter.stroke_opacity = ValueRef(value=0.5)

#add format to map
vis.rebind(column='Poverty Percent All Ages', brew='YlGnBu')
vis.scales['color'].type = 'threshold'
vis.scales['color'].domain = [0, 4, 6, 8, 10, 12, 20, 30]
vis.legend(title='Poverty 2016 (%)')
```

```python
#3. Ratio between percentage of change in poverty from 2000 to 2016 for each county. 
#Divided by the nation-wide change in poverty (find this value).

#define range of years
from_ = '2000'
to_ = '2016'

num = float(US_stat['Poverty Percent All Ages'][US_stat.year == to_])
denom = float(US_stat['Poverty Percent All Ages'][US_stat.year == from_])
perc_change_poverty_nation = round(num/denom,10)

merged_perc_change = []
for county in np.unique(merged.FIPS):
    
    try:
        #[poverty % in 2016] / [poverty % in 2000]
        #compute the equation above
        num = float(merged['Poverty Percent All Ages'][(merged.year == to_) & (merged.FIPS == county)])
        denom = float(merged['Poverty Percent All Ages'][(merged.year == from_) & (merged.FIPS == county)])
        perc_change_poverty = round(num/denom,10)

        perc_change_ratio = perc_change_poverty/perc_change_poverty_nation
    except:
        perc_change_ratio = np.nan        

    merged_perc_change.append((county,perc_change_ratio))

df_perc_change_ratio = pd.DataFrame(merged_perc_change,columns=['FIPS','Percent Change in Poverty Nation-Wide Ratio'])
df_perc_change_ratio.head()
```

```python
#check the min and max values to define the color scale
print('Min: {}'.format(min(df_perc_change_ratio['Percent Change in Poverty Nation-Wide Ratio'])))
print('Max: {}'.format(max(df_perc_change_ratio['Percent Change in Poverty Nation-Wide Ratio'])))
```

```python
from vincent import ValueRef
vincent.core.initialize_notebook()

#set-up the map
county_topo = r'us_counties.topo.json'
geo_data = [{'name': 'counties',
             'url': county_topo,
             'feature': 'us_counties.geo'}]

#plot the map
vis = vincent.Map(data=df_perc_change_ratio, geo_data=geo_data, scale=1100,
                  projection='albersUsa', data_bind='Percent Change in Poverty Nation-Wide Ratio',
                  data_key='FIPS', map_key={'counties': 'properties.FIPS'})
vis.marks[0].properties.enter.stroke_opacity = ValueRef(value=0.5)

#add format to map
vis.rebind(column='Percent Change in Poverty Nation-Wide Ratio', brew='YlGnBu')
vis.scales['color'].type = 'threshold'
vis.scales['color'].domain = [0,0.3,0.6,0.9,1.2,1.5,1.8]
vis.legend(title='Percent Change in Poverty Nation-Wide Ratio')
```

### Q5: Identify the extremes.
Find out which are the counties in which the poverty percentage increased or decreased the most during the period 2000 - 2016.

```python
# Do not run this cell
# If you do analysis for 2000-2013, Answer looks like some thing below
```

```python
#define range of years
from_ = '2000'
to_ = '2016'

#copy county_stat data and remove multilevel index
df = county_stat.reset_index()

#calculate the percent change nation-wide
num = float(US_stat['Poverty Percent All Ages'][US_stat.year == to_])
denom = float(US_stat['Poverty Percent All Ages'][US_stat.year == from_])
perc_change_poverty_nation = num/denom

#loop over every county to find percent change, the normalize it to nation-wide change
merged_perc_change = []
for fips in np.unique(df.FIPS):
    
    try:
        #[poverty % in 2016] / [poverty % in 2000]
        #compute the equation above
        num = float(df['Poverty Percent All Ages'][(df.year == to_) & (df.FIPS == fips)])
        denom = float(df['Poverty Percent All Ages'][(df.year == from_) & (df.FIPS == fips)])
        perc_change_poverty = num/denom

        perc_change_ratio = perc_change_poverty/perc_change_poverty_nation
    except:
        perc_change_ratio = np.nan #assign nan when data is not present
        
    #state and county
    county = np.unique(df.Name[df.FIPS == fips])[0]
    state = np.unique(df.Postal[df.FIPS == fips])[0]

    #append data as tuples
    merged_perc_change.append((fips,county,state,perc_change_ratio))

#create dataframe
df_perc_change_ratio = pd.DataFrame(merged_perc_change,columns=['FIPS','County','State','Percent Change in Poverty Nation-Wide Ratio'])
df_perc_change_ratio.head()
```

```python
#find max increase and max decrease
max_increase = df_perc_change_ratio.loc[df_perc_change_ratio['Percent Change in Poverty Nation-Wide Ratio'].idxmax()]
max_decrease = df_perc_change_ratio.loc[df_perc_change_ratio['Percent Change in Poverty Nation-Wide Ratio'].idxmin()]

#print results
print('County with max increase in poverty -- {} in state {}'.format(max_increase['County'].strip(),max_increase['State']))
print('change = {}'.format(max_increase['Percent Change in Poverty Nation-Wide Ratio']))

print('County with max decrease in poverty -- {} in state {}'.format(max_decrease['County'].strip(),max_decrease['State']))
print('change = {}'.format(max_decrease['Percent Change in Poverty Nation-Wide Ratio']))
```
